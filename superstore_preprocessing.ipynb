```python
import pandas as pd

# 加载原始数据
file_path = '/Users/caishiyu/Desktop/superstore-bi-dashboard/Sample - Superstore.csv'
df = pd.read_csv(file_path, encoding="ISO-8859-1")
df.head()
```

```python
# 查看数据结构与字段类型
df.info()
```

```python
# 描述统计（包含所有字段）
df.describe(include='all')
```

```python
# 每列缺失值数量
df.isnull().sum()
```

```python
# 将日期字段转换为 datetime 类型
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# 新增年月字段，便于后续按时间分析
df['Year-Month'] = df['Order Date'].dt.to_period('M')
df[['Order Date', 'Year-Month']].head()
```

```python
# 构造利润率字段（避免除以 0）
df['Profit Rate'] = df['Profit'] / df['Sales']
df['Profit Rate'] = df['Profit Rate'].replace([float('inf'), -float('inf')], 0)
df[['Sales', 'Profit', 'Profit Rate']].head()
```

```python
# 删除重复值
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. 销售趋势分析
monthly_sales = df.groupby('Year-Month')['Sales'].sum().reset_index()
monthly_sales['Year-Month'] = monthly_sales['Year-Month'].astype(str)
monthly_sales
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_sales, x='Year-Month', y='Sales', marker='o')
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("monthly_sales_trend.png")
plt.show()
```

```python
# 2. 地区销售额对比
region_sales = df.groupby('Region')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
region_sales
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(data=region_sales, x='Region', y='Sales', hue='Region', palette='Blues_d', legend=False)
plt.title("Total Sales by Region")
plt.tight_layout()
plt.savefig("region_sales_bar.png")
plt.show()
```

```python
# 3. 品类利润分析
category_profit = df.groupby('Category')['Profit'].sum().reset_index().sort_values(by='Profit')
category_profit
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=category_profit, x='Profit', y='Category', hue='Category', palette='Set2', legend=False)
plt.title("Profit by Category")
plt.tight_layout()
plt.savefig("category_profit_bar.png")
plt.show()
```

```python
# 4. 品类销售分析
category_sales = df.groupby('Category')['Sales'].sum().reset_index().sort_values(by='Sales')
category_sales
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=category_sales, x='Sales', y='Category', hue='Category', palette='Set2', legend=False)
plt.title("Sales by Category")
plt.tight_layout()
plt.savefig("category_sales_bar.png")
plt.show()
```

```python
# 5. 子类销售额前 10
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False).reset_index(drop=True).head(10)
subcat_sales
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=subcat_sales, x='Sales', y='Sub-Category', hue='Sub-Category', palette='viridis', legend=False)
plt.title("Top 10 Sub-Category Sales")
plt.tight_layout()
plt.savefig("subcat_sales_top10.png")
plt.show()
```

```python
# 新增发货天数字段
df['Ship Duration'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Ship Duration']
```

```python
# 保存清洗好的数据
df.to_csv("superstore_cleaned.csv", index=False)
print("✅ 数据清洗完成并保存为 superstore_cleaned.csv")
```
